```python
def drop_missing_value(df, threshold=0.6):
    missing_ratio = df.isnull().mean()
    columns_to_drop = missing_ratio[missing_ratio > threshold].index
    df.drop(columns=columns_to_drop,inplace=True)
    return df
```

# WoE và IV

```python
def WoE_categorical(df,feature):
    df[feature+'_WoE']=df[feature].map(
        np.log(
            ((df.groupby(feature).TARGET.count()-df.groupby(feature).TARGET.sum())/(df.TARGET.count()-df.TARGET.sum()))/(
            df.groupby(feature).TARGET.sum()/df.TARGET.sum()
            )
        )
    )
    df[feature+'_WoE']=df[feature+'_WoE'].to_list()
    df.replace([np.inf, -np.inf], 0, inplace=True)
    return df

#IV=WoE*(good loan%-bad loan%)

def IV_categorical(df,features):
    IV_table={}
    for item in features:
        All_Good=df.TARGET.count()-df.TARGET.sum()
        All_Bad=df.TARGET.sum()
        IV_table[item]=(df.groupby(item)[item+'_WoE'].mean()*(
            ((df.groupby(item).TARGET.count()-df.groupby(item).TARGET.sum())/All_Good-df.groupby(item).TARGET.sum()/All_Bad)
        )).sum().round(8)
    IV_table=pd.DataFrame.from_dict(IV_table,orient='index',columns=['IV']).reset_index().rename(columns={'index':'Features'}).sort_values(by='IV')
    return IV_table
```

```python
category_column = data.columns[data.dtypes == 'object']
number_column = data.columns[data.dtypes != 'object']
len(number_column), len(category_column)
```

```python
all_cat_col=category_column.to_list()
```

```python
data_WoE = data.copy()
for item in all_cat_col:
    data_WoE = WoE_categorical(data_WoE, item)
IV_table = IV_categorical(data_WoE, all_cat_col)
```

```python
all_cont_col = number_column
df = data.copy()
bins_col = []

for item in all_cont_col:
    # Phân nhóm dữ liệu cho các cột số thành 10 nhóm bằng phương pháp qcut
    df[item + '_bins'] = pd.qcut(data[item], 10, duplicates='drop')
    # Áp dụng WoE cho các nhóm đã phân
    df = WoE_categorical(df, item + '_bins')
    # Lưu tên cột bins vào danh sách
    bins_col.append(item + '_bins')
```

# Xử lý mất cân bằng dữ liệu

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import SMOTE

# Kiểm tra tỷ lệ ban đầu
print("Tỷ lệ ban đầu:")
print(y_train.value_counts())

# Bước 1: Undersampling lớp chiếm đa số (0) để đạt tỷ lệ 3:1
count_class_1 = y_train.value_counts()[1]  # Số lượng lớp thiểu số
count_class_0 = y_train.value_counts()[0]  # Số lượng lớp chiếm đa số

# Đặt số lượng lớp chiếm đa số (class 0) sao cho tỷ lệ là 3:1
count_class_0 = min(count_class_0, count_class_1 * 3)  # Đảm bảo không vượt quá số lượng của lớp chiếm đa số

# Thực hiện Undersampling
rus = RandomUnderSampler(sampling_strategy={0: count_class_0, 1: count_class_1}, random_state=42)
X_under, y_under = rus.fit_resample(x_train, y_train)

# Kiểm tra lại tỷ lệ sau khi Undersampling
print("\nSau khi Undersampling:")
print(y_under.value_counts())

# Bước 2: Oversampling lớp thiểu số với SMOTE để giữ tỷ lệ 3:1
smote = SMOTE(sampling_strategy=0.5, random_state=42)  # Tăng lớp thiểu số để giữ lại tỷ lệ ban đầu
X_balanced_train, y_balanced_train = smote.fit_resample(X_under, y_under)

# Kiểm tra lại tỷ lệ sau khi kết hợp cả hai kỹ thuật
print("\nSau khi Oversampling với SMOTE:")
print(y_balanced_train.value_counts())

# Kết hợp lại dữ liệu đã cân bằng
balanced_application_train = pd.concat([pd.DataFrame(X_balanced_train), pd.DataFrame(y_balanced_train, columns=['TARGET'])], axis=1)
```

# Chạy mô hình

```python
from sklearn.preprocessing import StandardScaler

X_train = balanced_application_train.drop(columns=['TARGET'])
y_train = balanced_application_train['TARGET']

# Chuẩn hóa dữ liệu (scaling)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(x_test)  # Dữ liệu kiểm tra được chuẩn hóa cùng với huấn luyện
```

```python
X_train.shape, y_train.shape
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, roc_auc_score, f1_score
import numpy as np

# Huấn luyện mô hình Logistic Regression
model_lr = LogisticRegression(random_state=42)
model_lr.fit(X_train_scaled, y_train)

# Dự đoán xác suất cho lớp 1
y_pred_proba = model_lr.predict_proba(X_test_scaled)[:, 1]

# Tìm ngưỡng tối ưu cho F1-score
thresholds = np.arange(0.0, 1.1, 0.01)
f1_scores = []

# Tính F1-score cho từng ngưỡng
for threshold in thresholds:
    y_pred = (y_pred_proba >= threshold).astype(int)
    f1 = f1_score(y_test, y_pred)
    f1_scores.append(f1)

# Chọn ngưỡng tối ưu (ngưỡng có F1-score cao nhất)
optimal_threshold = thresholds[np.argmax(f1_scores)]
print(f"Optimal threshold for F1-score: {optimal_threshold:.4f}")
print(f"Max F1-score: {max(f1_scores):.4f}")

# Dự đoán với ngưỡng tối ưu
y_pred_optimal = (y_pred_proba >= optimal_threshold).astype(int)

# In các chỉ số đánh giá mô hình
print("Classification Report:")
print(classification_report(y_test, y_pred_optimal))

# Vẽ ma trận nhầm lẫn
cm = confusion_matrix(y_test, y_pred_optimal)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='g', cmap='Blues', xticklabels=['Predicted 0', 'Predicted 1'], yticklabels=['Actual 0', 'Actual 1'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title(f'Confusion Matrix (Optimal Threshold: {optimal_threshold:.4f})')
plt.show()
```
